# Distribution of Cars by Fuel Type

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Fuel_Type', data=data)
plt.title('Distribution of Cars by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Cars')
plt.show()
    
    
```

# Number of Cars by Transmission Type


```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Transmission', data=data)
plt.title('Number of Cars by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Number of Cars')
plt.show()
```

# Price Distribution by Year


```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Present_Price', data=data, label='Present Price')
sns.lineplot(x='Year', y='Selling_Price', data=data, label='Selling Price')
plt.title('Price Distribution by Year')
plt.xlabel('Year')
plt.ylabel('Price (in lakhs)')
plt.legend()
plt.show()
```

# Kilometers Driven vs. Year


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Year', y='Kms_Driven', data=data)
plt.title('Kilometers Driven vs. Year')
plt.xlabel('Year')
plt.ylabel('Kilometers Driven')
plt.show()
```

# Ownership Status Distribution



```python
# Ownership Status Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Owner', data=data)
plt.title('Ownership Status Distribution')
plt.xlabel('Ownership Status')
plt.ylabel('Number of Cars')
plt.show()
```

# Price Comparison by Seller Type

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Seller_Type', y='Present_Price', data=data)
plt.title('Present Price Comparison by Seller Type')
plt.xlabel('Seller Type')
plt.ylabel('Present Price (in lakhs)')
plt.show()
```

# Selling Price Comparison by Seller Type

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Seller_Type', y='Selling_Price', data=data)
plt.title('Selling Price Comparison by Seller Type')
plt.xlabel('Seller Type')
plt.ylabel('Selling Price (in lakhs)')
plt.show()
```

# Present Price vs Selling Price

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Present_Price', y='Selling_Price', data=data)
plt.title('Present Price vs Selling Price')
plt.xlabel('Present Price (in lakhs)')
plt.ylabel('Selling Price (in lakhs)')
plt.show()
```
